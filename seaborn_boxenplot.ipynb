###boxenplot
- 더 큰 데이터 세트를 위한 향상된 상자 플롯


###boxenplot 실습에 사용할 데이터
- tips
 - total_bill(float64) : 최종 금액
 - tip(float64) : 팁
 - sex(category) : 성별
 - smoker(category) : 흡연자/비흡연자
 - day(category) : 요일(Sun/Sat/Thur/Fri)
 - time(category) : 시간(Dinner/Lunch)
 - size(int64) : 인원(1~6)

####그래프 확인

```python
#volinplot, boxplot 과 비교하며 보기

f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x=tips['total_bill'],ax=ax[0])
sns.violinplot(x=tips['total_bill'], ax=ax[1])
sns.boxplot(x=tips['total_bill'], ax=ax[2])
plt.show()
```

```python
f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x='day', y='total_bill',data=tips, ax=ax[0])
sns.violinplot(x='day', y='total_bill',data=tips, ax=ax[1])
sns.boxplot(x='day', y='total_bill',data=tips, ax=ax[2])
```

```python
f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x='day',y='total_bill', data=tips, hue='smoker', ax=ax[0])
sns.violinplot(x='day',y='total_bill', data=tips, hue='smoker', ax=ax[1])
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker', ax=ax[2])
```

```python
f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x='day',y='total_bill', data=tips, hue='time', ax=ax[0],linewidth=2.5)
sns.violinplot(x='day',y='total_bill', data=tips, hue='time', ax=ax[1],linewidth=2.5)
sns.boxplot(x='day',y='total_bill', data=tips, hue='time', ax=ax[2],linewidth=2.5)
```

```python
f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x='time',y='tip', data=tips, order=["Dinner", "Lunch"], ax=ax[0],)
sns.violinplot(x='time',y='tip', data=tips, order=["Dinner", "Lunch"], ax=ax[1],)
sns.boxplot(x='time',y='tip', data=tips, order=["Dinner", "Lunch"], ax=ax[2],)
```

```python
f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x='total_bill', data=tips, ax=ax[0], orient='h')
sns.violinplot(x='total_bill', data=tips, ax=ax[1],orient='h')
sns.boxplot(x='total_bill', data=tips, ax=ax[2],orient='h')
```

```python
f, ax = plt.subplots(3,1, figsize=(10,10))
sns.boxenplot(x='day', y='total_bill', data=tips,showfliers=False, ax=ax[0])
sns.stripplot(x='day', y='total_bill', data=tips,color='.26', ax=ax[0])
sns.violinplot(x='day', y='total_bill', data=tips,showfliers=False, ax=ax[1])
sns.stripplot(x='day', y='total_bill', data=tips,color='.26', ax=ax[1])
sns.boxplot(x='day', y='total_bill', data=tips,showfliers=False, ax=ax[2])
sns.stripplot(x='day', y='total_bill', data=tips,color='.26', ax=ax[2])
```

```python
sns.catplot(x='sex', y='total_bill', hue='smoker',col='time',
            data=tips, kind='boxen', height=4, aspect=.7)
sns.catplot(x='sex', y='total_bill', hue='smoker',col='time',
            data=tips, kind='violin', height=4, aspect=.7)
sns.catplot(x='sex', y='total_bill', hue='smoker',col='time',
            data=tips, kind='box', height=4, aspect=.7)
```
