###relplot 실습에 사용할 데이터
- tips
- fmri

#### kind = scatter

```python
tips = sns.load_dataset('tips')
```

```python
#tips 데이터 상위 5개 
tips.head()
```

```python
#tips 데이터의 정보
tips.info()
```

```python
#tips 데이터의 통계적 요약
tips.describe()
```

#####relplot 그리기

```python
#kind 옵션의 default는 scatter 이다.
#x축은 total_bill(float64), y축은 tip(float64)
relplot = sns.relplot(data=tips, x='total_bill',y='tip')
```

```python
#x축은 total_bill(float64), y축은 tip(float64)
#hue 옵션을 사용하기 - day(category=범주형)
relplot = sns.relplot(data=tips, x='total_bill',y='tip',hue='day')
```

```python
#x축은 total_bill(float64), y축은 tip(float64)
#col옵션을 사용하기 - time(category=범주형)
relplot = sns.relplot(data=tips, x='total_bill',y='tip',col='time')
```

```python
#x축은 total_bill(float64), y축은 tip(float64)
#hue옵션과 col옵션 함께 사용하기
#hue = day(category=범주형), col =  time(category=범주형)
relplot = sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='day')
```

```python
#x축은 total_bill(float64), y축은 tip(float64)
#row옵션 사용하기 - sex(category=범주형)
relplot = sns.relplot(data=tips, x='total_bill',y='tip',row='sex')
```

```python
#hue,col, row옵션 함께 사용하기
#hue, col, row 옵션에 사용할 변수는 category(변수형)을 사용
relplot = sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='day',row='sex')
```

```python
#col_wrap 옵션 사용하기
#col_wrap옵션은 col의 최대갯수를 몇개로 표시할 것인지
#col_wrap = 2  --> 최대 2개
relplot = sns.relplot(data=tips, x="total_bill", y="tip", hue="time", col="day",col_wrap=2)
```

####kind = line

```python
fmri = sns.load_dataset('fmri')
```

```python
#fmri 데이터 상위 5개 
fmri.head()
```

```python
#fmri 데이터의 정보
fmri.info()
```

```python
#fmri 데이터의 통계적 요약
fmri.describe()
```

#####relplot 그리기

```python
#x축 : timepoint(int64), y축 : signal(float64)
#kind='line' 옵션을 사용하여 lineplot으로 그림
relplot = sns.relplot(kind='line',data=fmri, x='timepoint',y='signal')
```

```python
#col옵션 - region(object)
#hue옵션 - event(object)
#region = 부위, parietal = 정수리, frontal = 정면
relplot = sns.relplot(kind='line',data=fmri,x='timepoint',y='signal', hue='event',col='region')
```

```python
#style 변경
relplot = sns.relplot(kind='line',data=fmri,x='timepoint',y='signal', hue='event',col='region',style='event')
```

```python
#set_axis_labels(x축, y축)은 x축의 이름, y축의 이름을 지정해줌
relplot = sns.relplot(kind='line',data=fmri,x='timepoint',y='signal', hue='event',col='region',style='event')
relplot.set_axis_labels("Timepoint", "Percent signal change")
relplot.set_titles("Region: {col_name} cortex")
```
