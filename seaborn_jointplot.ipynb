```python
#라이브러리 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import set_matplotlib_formats
```

###jointplot
- 산점도를 기본으로 표시, x, y축에 각 변수에 대한 히스트로그램을 동시에 보여줌
- 이변량, 일변량 그래프를 사용하여 두 변수의 플롯을 그림

###jointplot 실습에 사용할 데이터
- penguins
  - species(object) : 펭귄의 종
  - island(object) : 섬
  - bill_length_mm(float64) : 부리의 길이
  - bill_depth_mm(float64) : 부리의 깊이
  - flipper_length_mm(float64) : 팔의 길이
  - body_mass_g(float64) : 몸의 질량
  - sex(object) : 성별

####penguins

```python
#데이터 로드하기
penguins = sns.load_dataset('penguins')
```

```python
#데이터 정보
penguins.info()
```

```python
#데이터 수치적 요약
penguins.describe(include=np.number)
```

```python
#데이터 범주형 요약
penguins.describe(include='object')
```

```python
#결측치 확인
penguins[penguins['sex'].isnull()]
```

####그래프 확인

```python
sns.jointplot(data=penguins, x='bill_length_mm', y ="bill_depth_mm")
```

```python
#hue 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue="species")
```

```python
#kind = 'kde' 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', kind='kde')
```

```python
#kind='reg' 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='reg')
```

```python
#kind='hist 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='hist')
```

```python
#kind='hex' 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='hex')
```

```python
#marker = '+', s=100, marginal_kws = dict(bins=25, fill=False) 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', 
              marker='+', s=100, marginal_kws = dict(bins=25, fill=False))
```

```python
#height=5 , ratio = 2,  marginal_ticks =True, 옵션 사용
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', height=5, ratio=2, marginal_ticks=True)
```

```python
#return JointGrid

g = sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
g.plot_joint(sns.kdeplot, color='r', zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color='r', height=-.15, clip_on=False)
```
