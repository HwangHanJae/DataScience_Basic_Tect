```python
#데이터 정보
penguins.info()
```

```python
#데이터 수치적 요약
penguins.describe(include=np.number)
```

####그래프 확인

```python
#히스토그램
sns.histplot(data=penguins, x='flipper_length_mm')
```

```python
#x축과 y축 바꾸기
sns.histplot(data=penguins, y='flipper_length_mm')
```

```python
#binwidth=3 옵션 사용
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=3)
```

```python
#bins 옵션 사용
sns.histplot(data=penguins, x='flipper_length_mm', bins=30)
```

```python
#kde=True 옵션
#kde(커널 밀도 함수)를 그래프에 표시
sns.histplot(data=penguins, x='flipper_length_mm', kde=True)
```

```python
#x축과 y축에 할당하지 않은 경우
#데이터 세트는 와이드 형식으로 처리
#각 숫자열에 대해 히스토그램이 그려짐
sns.histplot(data=penguins)
```

```python
#hue 옵션 사용
#species(종)을 색별로 표현
sns.histplot(data=penguins, x='flipper_length_mm', hue='species')
```

```python
#multiple 옵션을 사용하여 변경 - layer(기본값)
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='layer')
```

```python
#multiple 옵션을 사용하여 변경 - stack
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='stack')
```

```python
#multiple 옵션을 사용하여 변경 - fill
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='fill')
```

```python
#multiple 옵션을 사용하여 변경 - dodge
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', multiple='dodge')
```

```python
#element='bars'옵션사용 - bars(기본값)
#겹치는 막대를 시각적으로 표현하기 위해
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', element='bars')
```

```python
#element='step'옵션사용 - step(기본값)
#겹치는 막대를 시각적으로 표현하기 위해
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', element='step')
```

```python
#element='poly'옵션사용 - poly(기본값)
#겹치는 막대를 시각적으로 표현하기 위해
#주의 : 히스토그램의 모양은 아니지만 분포를 더 쉽게 볼 수 있음 
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', element='poly')
```

```python
#크기가 다른 부분 집합의 분포를 비교하려면 독립 밀도 정규화를 사용
#stat='density', common_norm = False
sns.histplot(data = penguins, x='bill_length_mm', hue='island', element='step', stat='density', common_norm=False)
```

```python
#위의 그래프와 비교
sns.histplot(data = penguins, x='bill_length_mm', hue='island', stat='density', element='step')
```

```python
#x와 y가 모두 할당되면 이변량 히스토그램이 계산
#히트맵으로 표시
sns.histplot(data=penguins, x='bill_depth_mm', y='body_mass_g')
```

```python
#x와 y가 모두 할당되면 이변량 히스토그램이 계산
#hue 옵션 사용
#히트맵으로 표시
sns.histplot(data=penguins, x='bill_depth_mm', y='body_mass_g', hue='species')
```

```python
sns.histplot(
    penguins, x="bill_depth_mm", y="species", hue="species", legend=False
)
```

```python
#데이터 로드하기
tips = sns.load_dataset('tips')
```

```python
#데이터 정보
tips.info()
```

```python
#데이터 수치적 요약
tips.describe()
```

```python
#데이터 범주적 요약
tips.describe(include='category')
```

####그래프 확인

```python
#stat='probability'옵션 사용
#각 막대의 높이가 확률을 나타낼수록 정규화 할수 있음
sns.histplot(data=tips, x='size', stat='probability',discrete=True)
```

```python
#범주형 데이터 또한 히스토그램에 그릴 수 있음
#shrink 옵션 사용 - 막대 사이의 거리?(수축?)
sns.histplot(data=tips, x='day',shrink=0.8)
```

```python
#hue 옵션, multiple='dodge' 옵션 사용
sns.histplot(data=tips, x='day', hue='sex', multiple='dodge', shrink=0.8)
```

####planets

```python
#데이터 로드하기
planets = sns.load_dataset('planets')
```

```python
#데이터 정보
planets.info()
```

```python
#데이터 수치적 요약
planets.describe()
```

```python
#데이터 범주적 요약
planets.describe(include='object')
```

####그래프 확인

```python
sns.histplot(data=planets, x='distance')
```

```python
#log_scale=True 옵션 사용
#log화
sns.histplot(data=planets, x='distance', log_scale=True)
```

```python
#fill = False 옵션사용
sns.histplot(data=planets, x='distance', log_scale=True, fill = False)
```

```python
#element = 'step' 옵션
sns.histplot(data=planets, x='distance', log_scale=True, fill = False, element='step')
```

```python
#cumulative = True
#누적 사용
sns.histplot(
    data=planets, x='distance', hue='method',
    hue_order=['Radial Velocity', 'Transit'],
    log_scale = True, element='step',fill=False,
    stat='density', common_norm=False, cumulative =True
)
```

```python
#x축에는 year, y축에는 distance
#discrete(True, False), log_scale(False, True)
sns.histplot(data=planets, x='year', y='distance',
  bins=30, discrete=(True, False), log_scale=(False, True)
)
```

```python
 # thresh=None 옵션 사용
 sns.histplot(data=planets, x='year', y='distance',
  bins=30, discrete=(True, False), log_scale=(False, True),
  thresh=None
)
```

```python
sns.histplot(data=planets, x='year', y='distance',
             bins=30, discrete=(True, False), log_scale=(False, True),
             pthresh=.05, pmax=.9)
```

```python
#colorbar 옵션 사용
sns.histplot(data=planets, x='year', y='distance',
             bins=30, discrete=(True, False), log_scale=(False, True),
             cbar=True, cbar_kws=dict(shrink=.75))
```
