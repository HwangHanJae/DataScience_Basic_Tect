###clustermap - 클러스터맵
- 행렬 데이터 세트를 계층적으로 클러스터 된 히트맵으로 플로팅

###clustermap 실습에 사용할 데이터
- iris
 - sepal_length(float64) : 꽃받침 길이
 - sepal_width(float64) : 꽃받침 너비
 - petal_length(float64) : 꽃잎 길이
 - petal_width(float64) : 꽃잎 너비
 - species(object) : 종

####iris

```python
#데이터 로드하기
iris = sns.load_dataset('iris')
```

```python
#데이터 정보
iris.info()
```

```python
#데이터의 수치적 요약
iris.describe(include=np.number)
```

```python
#데이터 범주적 요약
iris.describe(include='object')
```

```python
species = iris.pop('species')
sns.clustermap(iris)
```

```python
#figsize, size를 변경
sns.clustermap(iris, figsize=(7,5), row_cluster=False,
               dendrogram_ratio=(.1, .2),
               cbar_pos=(0, .2, .03, .4))
```

```python
#색상 레이블을 추가
lut = dict(zip(species.unique(), "rbg"))
row_colors = species.map(lut)
sns.clustermap(iris, row_colors=row_colors)
```

```python
sns.clustermap(iris, cmap='mako', vmin=0, vmax=10)
```

```python
#metric = 'correlation' 옵션 사용
sns.clustermap(iris, metric='correlation')
```

```python
#method = 'single'
sns.clustermap(iris, method='single')
```

```python
#standard_scale = 1 옵션 사용
sns.clustermap(iris, standard_scale=1)
```

```python
#z_ccore = 0 옵션 사용
sns.clustermap(iris, z_score =0, cmap='vlag')
```
