###Categorical plots
- 숫자와 하나 이상의 범주 형 변수 간의 관계를 보여줌
- kind 매개 변수는 사용할 기본 축 수준 함수를 선택
- Categorical scatterplots:
  - stripplot (kind='strip' , default)
  - swarmplot (kind='swarm')
- Categorical distribution plots:
  - boxplot (kind='box)
  - violinplot (kind='violin')
  - boxenplot (kind='boxen')
- Categorical estimate plots:
  - pointplot (kind='point')
  - barplot (kind='bar')
  - countplot (kind='count')

###그래프 확인

```python
#x축은 time(category) y축은 pulse(int64)
#hue = 'kind' (category)
g = sns.catplot(data=exercise, x='time', y='pulse', hue='kind')
```

```python
#위의 그래프를 violinplot으로 변경
g = sns.catplot(data=exercise, x='time', y='pulse', hue='kind', kind='violin')
```

```python
#col 옵션 사용하여 플롯 그리기
g = sns.catplot(data=exercise, x='time', y='pulse', hue='kind', col='diet')
```

```python
#height , aspect를 변경하여 그래프 크기 변경
g = sns.catplot(data=exercise, x='time', y='pulse', hue='kind', col='diet', height=5, aspect=.8)
```

```python
g = sns.catplot(x='alive', col='deck', col_wrap=4, 
                data=titanic[titanic['deck'].notnull()],
                kind='count', height=2.5, aspect=.8)
```

```python
g = sns.catplot(x='age', y='embark_town', hue='sex', row='class',
                data=titanic[titanic['embark_town'].notnull()],
                orient="h",
                             height=2, aspect=3,
                             kind='violin', dodge=True, cut=0, bw=.2)
```

```python
g = sns.catplot(x='who', y='survived', col='class',
                data=titanic, kind='bar', ci=None, aspect=.6, saturation=.5)
g.set_axis_labels('', 'Survived Rate')
g.set_xticklabels(["Men", "Women","Children"])
g.set_titles("{col_name} {col_var}")
g.set(ylim=(0,1))
g.despine(left=True)
```
