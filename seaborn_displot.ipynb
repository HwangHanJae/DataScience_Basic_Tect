###seaborn 라이브러리 정리
- [seaborn API 링크](https://seaborn.pydata.org/api.html)
- seborn은 파이썬 데이터 시각화 라이브러리
- matplotlib에 기반하였음

```python
#라이브러리 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import set_matplotlib_formats
```

```python
#그래프가 잘보이게 하기 위해 'ggplot' 스타일을 사용
plt.style.use('ggplot')
#폰트가 선명하게 보이게 하기위하여
set_matplotlib_formats('svg')
```

###Distribution plots - 분포도
- displot
  - kind = 'hist' (히스토그램)
  - kind = 'kde' (커널 밀도 함수)
  - kind = 'ecdf' (경험적 누적 분포 함수)

###displot 실습에 사용할 데이터
- penguins
  - species(object) : 펭귄의 종
  - island(object) : 섬
  - bill_length_mm(float64) : 부리의 길이
  - bill_depth_mm(float64) : 부리의 깊이
  - flipper_length_mm(float64) : 팔의 길이
  - body_mass_g(float64) : 몸의 질량
  - sex(object) : 성별

```python
#데이터 로드하기
penguins = sns.load_dataset("penguins")
```

```python
#데이터 정보
penguins.info()
```

```python
#수치형 데이터의 통계적 요약
penguins.describe()
```

```python
#범주형 데이터 요약
penguins.describe(include='object')
```

```python
#결측치 확인
penguins[penguins['sex'].isnull()]
```

###그래프 확인

```python
#displot의 default plot의 종류는 히스토그램이다.
#x='flipper_length_mm', y='count'
sns.displot(data=penguins, x='flipper_length_mm')
```

```python
#kind='kde'옵션 사용
#커널 밀도 추정(Kernel Density Estimator)
sns.displot(data=penguins, x='flipper_length_mm', kind='kde')
```

```python
#kind=ecdf
#경험적 누적 분포 함수
sns.displot(data=penguins, x='flipper_length_mm', kind='ecdf')
```

```python
#kde=True 옵션
#히스토그램과 커널 밀도 함수(kde)함께 표시
sns.displot(data=penguins, x='flipper_length_mm', kde=True)
```

```python
#x값과 y값을 할당하면 bivariateplot(이변량플롯)을 그림
sns.displot(data=penguins, x='flipper_length_mm', y='bill_length_mm')
```

```python
#bivariateplot(이변량플롯)은 히스토그램, kde(커널 밀도 함수)에만 사용할 수 있음
sns.displot(data=penguins, x='flipper_length_mm', y='bill_length_mm', kind='kde')
```

```python
#rug 옵션
#개별 관측치를 표시할 수 있음
sns.displot(data=penguins,x='flipper_length_mm', y='bill_length_mm',rug=True)
```

```python
#hue옵션 사용
#species(종)별로 다른 색상 적용
sns.displot(data=penguins, x='flipper_length_mm', hue='species')
```

```python
#hue옵션 사용
#kind=kde 옵션 사용
#species(종)별로 다른 색상 적용
sns.displot(data=penguins, x='flipper_length_mm', hue='species', kind='kde')
```

```python
#multiple 옵션 사용
#mulitple 옵션의 defalut값은 layer
sns.displot(data=penguins, x='flipper_length_mm', hue='species', multiple="layer")
```

```python
#multiple=stack 옵션
sns.displot(data=penguins, x='flipper_length_mm', hue='species', multiple="stack")
```

```python
#multiple=fill 옵션
sns.displot(data=penguins, x='flipper_length_mm', hue='species', multiple="fill")
```

```python
#multiple=dodge 옵션
sns.displot(data=penguins, x='flipper_length_mm', hue='species', multiple="dodge")
```

```python
#col 옵션 사용하여 서브 플롯 표현
#col = 'sex' 서브플롯을 성별로 나누어 표현
sns.displot(data=penguins, x='flipper_length_mm', hue='species', kind='kde',col='sex')
```

```python
#height과 aspect 옵션으로 크기 조절
sns.displot(data=penguins, x='flipper_length_mm', hue='sex', kind='kde',col='species',height=4, aspect=.7)
```

```python
#displot() 함수는 FacetGrid 타입으로 리턴
#FacetGrid의 함수를 사용할 수 있음
#set_axis_labels(x,y)함수는 x,y의 이름을 변경할 수 있음
#set_titles()함수는 제목을 바꿀 수 있음
grid = sns.displot(data=penguins, x='flipper_length_mm', hue='sex', col='species', kind='kde', aspect=.6)
grid.set_axis_labels('Flipper length(mm)', '')
grid.set_titles('{col_name}')
```
